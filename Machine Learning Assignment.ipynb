#### Q1: Explain the following with an example:

**A) Artificial Intelligence (AI):**  
AI is a broad field of computer science that focuses on creating systems capable of performing tasks that usually require human intelligence. Examples include decision-making, speech recognition, visual perception, and language translation.

*Example:*  
A chatbot that can answer customer questions or a virtual assistant like Siri or Alexa is an example of AI. These systems use natural language processing to understand and respond to human queries.

**B) Machine Learning (ML):**  
ML is a subset of AI that involves teaching computers to learn from data without explicitly programming them. It focuses on developing algorithms that can learn patterns from data and make predictions or decisions based on it.

*Example:*  
A spam filter that learns to distinguish between spam and non-spam emails by analyzing various email features like text content, sender, and subject is an example of ML.

**C) Deep Learning (DL):**  
DL is a specialized subset of ML that uses neural networks with many layers (deep neural networks) to learn from vast amounts of data. It is particularly useful in tasks where the input data is highly complex, like image and speech recognition.

*Example:*  
Image recognition systems, such as those used by self-driving cars to identify objects like pedestrians, traffic lights, and other vehicles, use deep learning.

#### Q2: What is supervised learning? List some examples of supervised learning.

**Supervised Learning:**  
Supervised learning is a type of machine learning where the model is trained on a labeled dataset, which means each training example is paired with an output label. The model learns to map inputs to the correct output, making it useful for tasks where the output is known.

*Examples of Supervised Learning:*

- Linear Regression
- Logistic Regression
- Decision Trees
- Support Vector Machines (SVM)
- K-Nearest Neighbors (KNN)

#### Q3: What is unsupervised learning? List some examples of unsupervised learning.

**Unsupervised Learning:**  
Unsupervised learning is a type of machine learning where the model is trained on an unlabeled dataset. The model learns the patterns and structures in the input data without any explicit guidance on what to predict. It is often used for clustering and association tasks.

*Examples of Unsupervised Learning:*

- K-Means Clustering
- Hierarchical Clustering
- Principal Component Analysis (PCA)
- Independent Component Analysis (ICA)
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

#### Q4: What is the difference between AI, ML, DL, and DS?

- **Artificial Intelligence (AI):** AI is the overarching field that encompasses all techniques and methods to make machines intelligent.
- **Machine Learning (ML):** ML is a subset of AI that focuses on developing algorithms that allow computers to learn from and make predictions based on data.
- **Deep Learning (DL):** DL is a specialized subfield of ML that involves neural networks with many layers (deep architectures) and is particularly powerful for analyzing large, complex datasets.
- **Data Science (DS):** DS is a multidisciplinary field that involves extracting insights and knowledge from data. It includes statistics, data analysis, machine learning, and domain expertise.

#### Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?

- **Supervised Learning:** Requires labeled data. The model learns to map inputs to specific outputs.
- **Unsupervised Learning:** Does not require labeled data. The model identifies patterns and structures within the input data.
- **Semi-Supervised Learning:** Combines both labeled and unlabeled data. A small amount of labeled data is used to guide the learning, while a large amount of unlabeled data is used to improve the model's performance.

#### Q6: What is train, test, and validation split? Explain the importance of each term.

- **Training Set:** The portion of data used to train the model. The model learns from this data by adjusting its parameters to minimize the error.
- **Validation Set:** The portion of data used to evaluate the model during training. It helps in tuning model hyperparameters to avoid overfitting.
- **Test Set:** The portion of data used to evaluate the model's performance after training. It assesses the model's ability to generalize to new, unseen data.

*Importance:*

- **Training Set:** Helps the model learn.
- **Validation Set:** Guides the model tuning process.
- **Test Set:** Provides an unbiased evaluation of the final model.

#### Q7: How can unsupervised learning be used in anomaly detection?

Unsupervised learning can be used in anomaly detection by identifying patterns that deviate from the norm in the data. For example, clustering algorithms like K-Means can group similar data points together; data points that don't fit well into any cluster can be considered anomalies.
