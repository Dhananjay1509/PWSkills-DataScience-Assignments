Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### five types of plots that can be created using Seaborn, along with their uses:
#### 1. Scatter Plot
##### Use: Scatter plots are used to visualize the relationship between two continuous variables. They help in identifying correlations, trends, and outliers in the data.
##### Example: A scatter plot can show the relationship between petal length and petal width in different species of iris flowers.
#### 2. Line Plot
##### Use: Line plots are ideal for displaying trends over time or ordered categories. They connect individual data points with a line, making it easier to see changes across a continuous variable.
##### Example: A line plot can be used to visualize the trend of average temperatures over several years.
#### 3. Bar Plot
##### Use: Bar plots are used to compare the quantities of different categories. They represent the central tendency (e.g., mean) of a numerical variable for each category.
##### Example: A bar plot can show the average sales figures for different product categories.
#### 4. Heatmap
##### Use: Heatmaps visualize matrix-like data, where individual values are represented by colors. They are particularly useful for displaying correlations between variables.
##### Example: A heatmap can be used to show the correlation matrix of different features in a dataset, highlighting strong and weak correlations.
#### 5. Box Plot
##### Use: Box plots depict the distribution of a dataset through its quartiles, highlighting the median, range, and potential outliers. They are useful for comparing distributions across different groups.
##### Example: A box plot can illustrate the distribution of test scores across different classes, showing variations and outliers within each class.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot with x="timepoint" and y="signal" for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Add labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Signal over Time for Different Events and Regions")

# Show the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Create the first box plot for 'age'
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Age')

# Create the second box plot for 'fare'
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Fare')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a histogram for the 'price' column with hue for 'cut'
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=30, kde=True)

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices by Cut')

# Show the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a pair plot with hue for the 'species' column
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])

# Show the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a heatmap
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='g')

# Add labels and title
plt.title('Number of Passengers per Month Over Years')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
