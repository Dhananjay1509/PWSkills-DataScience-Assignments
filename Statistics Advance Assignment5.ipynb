#### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

#### Calculating a 95% Confidence Interval

To calculate a 95% confidence interval for a sample of data, we need the following information:

- **Sample mean (ùë•ÃÑ):** 50
- **Sample standard deviation (s):** 5
- **Sample size (n):** 30 (assuming)
- **Z-score corresponding to a 95% confidence level (Z):** 1.96

The formula for the confidence interval is:

$
CI = \bar{x} \pm Z \cdot \left( \frac{s}{\sqrt{n}} \right)
$

where:

- $\bar{x}$ is the sample mean,
- $s$ is the standard deviation,
- $n$ is the sample size,
- $Z$ is the Z-score corresponding to the desired confidence level (1.96 for a 95% confidence level).

#### Python Code

```python
import scipy.stats as stats

# Given data
mean = 50
std_dev = 5
sample_size = 30

# Z-score for 95% confidence level
z_score = stats.norm.ppf(0.975)  # 0.975 because it's a two-tailed test

# Standard error
standard_error = std_dev / (sample_size ** 0.5)

# Confidence interval
margin_of_error = z_score * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

confidence_interval
```

#### Interpretation
The 95% confidence interval calculated will give us a range within which the true population mean is expected to fall 95% of the time, given that our sample is representative of the population. For example, if the confidence interval were (48.21, 51.79), it would mean we are 95% confident that the true population mean lies between 48.21 and 51.79.

#### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

#### Chi-Square Goodness of Fit Test for M&M Colors

In this example, we'll conduct a Chi-square goodness of fit test to determine if the distribution of colors of M&Ms matches the expected distribution of:

- 20% Blue
- 20% Orange
- 20% Green
- 10% Yellow
- 10% Red
- 20% Brown

##### Step-by-Step Implementation

##### 1. Define the Observed and Expected Counts

Let's assume you have the following observed counts from a bag of M&Ms:

- Blue: 30
- Orange: 25
- Green: 28
- Yellow: 10
- Red: 12
- Brown: 25

Based on the expected distribution, if there are 100 M&Ms in a bag, the expected counts for each color would be:

- Blue: 20
- Orange: 20
- Green: 20
- Yellow: 10
- Red: 10
- Brown: 20

##### 2. Perform the Chi-Square Test using SciPy

You can perform the chi-square goodness of fit test using the `chisquare` function from the `scipy.stats` module. Here is the Python code to implement the test:

```python
import numpy as np
import scipy.stats as stats

# Observed counts (example values)
observed = np.array([30, 25, 28, 10, 12, 25])  # Adjust these as needed
# observed = np.array([20, 20, 20, 10, 10, 20])  # This sums to 100

# Expected counts based on the distribution for 100 M&Ms
expected = np.array([20, 20, 20, 10, 10, 20])

# Check if the sums match
if np.sum(observed) != np.sum(expected):
    print("Error: The sum of observed counts does not match the sum of expected counts.")
else:
    # Perform the chi-square goodness of fit test
    chi_square_statistic, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

    # Output the results
    print(f'Chi-Square Statistic: {chi_square_statistic}')
    print(f'P-Value: {p_value}')

    # Determine if we reject the null hypothesis
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The observed distribution does not match the expected distribution.")
    else:
        print("Fail to reject the null hypothesis: The observed distribution matches the expected distribution.")
```

#### Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following data:

|------------|---------|---------|
| **Outcome 1** | 20      | 15      |
| **Outcome 2** | 10      | 25      |
| **Outcome 3** | 15      | 20      |
#### Interpret the results of the test.

To calculate the chi-square statistic and p-value for a contingency table in Python, we can use the `chi2_contingency` function from the `scipy.stats` module. The contingency table data is represented as a 2D array.

Here is the Python code to perform the chi-square test:

```python
import numpy as np
from scipy.stats import chi2_contingency

# Define the contingency table
table = np.array([[20, 15],
                  [10, 25],
                  [15, 20]])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(table)

# Output the results
chi2, p, dof, expected
```

#### Interpretation of Results

- **Chi-square statistic (œá¬≤)**: The test statistic measures the difference between the observed and expected frequencies. A higher value indicates a greater difference.

- **p-value**: The probability of observing a chi-square statistic at least as extreme as the one computed, assuming that the null hypothesis (no association between the groups) is true. A low p-value (typically < 0.05) suggests rejecting the null hypothesis.

- **Degrees of freedom (dof)**: The number of categories minus 1 for rows and columns.

- **Expected frequencies**: The table of expected frequencies assuming no association between the groups.

By evaluating the p-value, you can decide whether to reject or accept the null hypothesis. If the p-value is less than 0.05, you would reject the null hypothesis, indicating a significant association between the groups and the outcomes. If it is greater than 0.05, you would not reject the null hypothesis, suggesting no significant association.


#### Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

To calculate the 95% confidence interval for the proportion of individuals who smoke in the population, you can use Python with the help of the scipy library. Here‚Äôs a step-by-step guide to calculate the confidence interval:

#### Calculate the sample proportion (ùëùÃÇ):

$
ùëùÃÇ = \frac{\text{number of smokers}}{\text{total number of individuals}} = \frac{60}{500} = 0.12
$

#### Calculate the standard error (SE) of the proportion:

$
SE = \sqrt{\frac{ùëùÃÇ (1 - ùëùÃÇ)}{ùëõ}}
$

where ùëõ is the sample size (500 in this case).

#### Determine the Z-score for a 95% confidence level. 
For a 95% confidence interval, the Z-score is approximately 1.96.

#### Calculate the confidence interval:

$
CI = ùëùÃÇ ¬± ùëç √ó SE
$


```python
import scipy.stats as stats
import numpy as np

# Given data
n = 500  # total number of individuals
x = 60   # number of smokers
p_hat = x / n  # sample proportion

# Standard error of the proportion
SE = np.sqrt(p_hat * (1 - p_hat) / n)

# Z-score for 95% confidence interval
z_score = stats.norm.ppf(0.975)

# Confidence interval calculation
margin_of_error = z_score * SE
ci_lower = p_hat - margin_of_error
ci_upper = p_hat + margin_of_error

print(f"95% confidence interval for the proportion of individuals who smoke: ({ci_lower:.4f}, {ci_upper:.4f})")
```

`np.sqrt` calculates the square root.  
`stats.norm.ppf(0.975)` gives the Z-score for the 95% confidence level.  
The margin of error is calculated as $ Z \times SE $, where:  
- $ Z $ is the Z-score for the desired confidence level
- $ SE $ is the standard error.

Finally, the lower and upper bounds of the confidence interval are found by subtracting and adding the margin of error from the sample proportion.

Running this code will give you the 95% confidence interval for the proportion of individuals who smoke in the population.


#### Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

To calculate a 90% confidence interval for a sample mean, you'll need to use the following formula:

$
CI = \text{{mean}} \pm \left( z \times \frac{{\text{{standard deviation}}}}{{\sqrt{n}}} \right)
$

where:

- **mean** is the sample mean,
- **z** is the z-score corresponding to the desired confidence level,
- **standard deviation** is the sample standard deviation,
- **n** is the sample size.

For a 90% confidence interval, the z-score is approximately 1.645. However, since the sample size $n$ isn't provided, I'll show you how to calculate the interval with a generic sample size. You can replace $n$ with your actual sample size.

Here's how you can calculate it in Python:


```python
import scipy.stats as stats
import math

# Given values
mean = 75
std_dev = 12
confidence = 0.90

# Assuming a sample size (for example, n = 30)
n = 30

# Calculate the standard error
std_error = std_dev / math.sqrt(n)

# Calculate the z-score for the 90% confidence interval
z_score = stats.norm.ppf(1 - (1 - confidence) / 2)

# Calculate the margin of error
margin_of_error = z_score * std_error

# Calculate the confidence interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print(f"90% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

#### Interpretation

The 90% confidence interval provides a range within which we expect the true population mean to lie with 90% confidence. If you use a sample size of 30, as in the example, the interval calculated will be around the sample mean of 75, adjusted by the margin of error. The interval will give you a sense of the precision of your estimate.


#### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

#### Plotting the Chi-Square Distribution with 10 Degrees of Freedom

To plot the chi-square distribution with 10 degrees of freedom and shade the area corresponding to a chi-square statistic of 15, you can use the `scipy` and `matplotlib` libraries in Python. 

##### Step-by-step code example:

1. **Import the necessary libraries**.
2. **Define the degrees of freedom and the chi-square statistic**.
3. **Generate the chi-square distribution data**.
4. **Plot the chi-square distribution**.
5. **Shade the area corresponding to the chi-square statistic**.

Here's a complete Python script to achieve this:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the degrees of freedom and the chi-square statistic
df = 10
chi_square_stat = 15

# Generate the chi-square distribution data
x = np.linspace(0, 30, 1000)
y = chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, label=f'Chi-Square Distribution (df={df})')
plt.fill_between(x, 0, y, where=(x >= chi_square_stat), color='red', alpha=0.5, label=f'Chi-Square Stat = {chi_square_stat}')

# Add labels and title
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

#### Explanation:
`np.linspace(0, 30, 1000)` generates an array of values from 0 to 30 (range for plotting).  
`chi2.pdf(x, df)` computes the probability density function for the chi-square distribution with `df` degrees of freedom.  
`plt.fill_between` is used to shade the area under the curve where the chi-square statistic is greater than 15.

#### Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

#### Calculating a 99% Confidence Interval for the True Proportion

To calculate a 99% confidence interval for the true proportion of people who prefer Coke, you can use the formula for the confidence interval of a proportion. Here's a step-by-step guide:

##### 1. Determine the sample proportion ($\hat{p}$):

$
\hat{p} = \frac{\text{Number of people who prefer Coke}}{\text{Total sample size}} = \frac{520}{1000} = 0.52
$

##### 2. Calculate the standard error (SE) for the proportion:

$
SE = \sqrt{\frac{\hat{p} \cdot (1 - \hat{p})}{n}} = \sqrt{\frac{0.52 \cdot (1 - 0.52)}{1000}} = \sqrt{\frac{0.52 \cdot 0.48}{1000}} = \sqrt{0.0002496} \approx 0.0158
$

##### 3. Find the z-value for a 99% confidence interval:

For a 99% confidence level, the z-value is approximately 2.576 (you can find this value from a z-table or standard normal distribution).

##### 4. Calculate the margin of error (ME):

$
ME = z \cdot SE = 2.576 \cdot 0.0158 \approx 0.0408
$

##### 5. Determine the confidence interval:

- **Lower bound**: 

$
\hat{p} - ME = 0.52 - 0.0408 = 0.4792
$

- **Upper bound**: 

$
\hat{p} + ME = 0.52 + 0.0408 = 0.5608
$

##### Conclusion:

So, the 99% confidence interval for the true proportion of people who prefer Coke is approximately:

$
[0.4792, 0.5608]
$



#### Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

#### Chi-Square Goodness of Fit Test: Coin Fairness

##### 1. State the Hypotheses
**Null Hypothesis (ùêª‚ÇÄ):** The coin is fair, meaning the probability of heads is 0.5 and the probability of tails is 0.5.  
**Alternative Hypothesis (ùêª‚Çê):** The coin is not fair, meaning the probabilities of heads and tails are not both 0.5.

##### 2. Calculate the Expected Frequencies
If the coin is fair, the expected frequencies for heads and tails in 100 flips are:

- Expected number of tails = 0.5 √ó 100 = 50  
- Expected number of heads = 0.5 √ó 100 = 50  

##### 3. Calculate the Chi-Square Test Statistic
The formula for the chi-square statistic is:

$
\chi^2 = \sum \frac{{(O_i - E_i)^2}}{{E_i}}
$

Where:  
- $O_i$ is the observed frequency  
- $E_i$ is the expected frequency  

Here:

- Observed number of tails ($O_{tails}$) = 45  
- Expected number of tails ($E_{tails}$) = 50  
- Observed number of heads ($O_{heads}$) = 100 - 45 = 55  
- Expected number of heads ($E_{heads}$) = 50  

Substitute these values into the formula:

$
\chi^2 = \frac{{(45 - 50)^2}}{{50}} + \frac{{(55 - 50)^2}}{{50}}
$

Calculate each term:

$
\frac{{(45 - 50)^2}}{{50}} = \frac{{(-5)^2}}{{50}} = \frac{{25}}{{50}} = 0.5
$

$
\frac{{(55 - 50)^2}}{{50}} = \frac{{5^2}}{{50}} = \frac{{25}}{{50}} = 0.5
$

Sum these values:

$
\chi^2 = 0.5 + 0.5 = 1.0
$

##### 4. Determine the Critical Value and Compare
For a chi-square test with 1 degree of freedom (df), the critical value at a 0.05 significance level can be found in chi-square distribution tables or using a chi-square calculator. The critical value is approximately **3.841**.

Compare the chi-square test statistic to the critical value:

$
\chi^2 = 1.0 < 3.841
$

##### 5. Make a Decision
Since the chi-square test statistic (1.0) is less than the critical value (3.841), we fail to reject the null hypothesis.

##### Conclusion
There is not enough evidence at the 0.05 significance level to conclude that the coin is biased towards tails. The observed frequencies are consistent with a fair coin.


#### Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis.

|                | Lung Cancer: Yes | Lung Cancer: No |
|----------------|------------------|-----------------|
| **Smoker**     | 60               | 140             |
| **Non-smoker** | 30               | 170             |

#### Use a significance level of 0.05.


#### Chi-Square Test for Independence

##### 1. Set Up the Hypotheses
- **Null Hypothesis (H‚ÇÄ):** Smoking status and lung cancer diagnosis are independent.
- **Alternative Hypothesis (H‚Çê):** Smoking status and lung cancer diagnosis are not independent.

##### 2. Construct the Contingency Table
Here is the provided table:

|                     | Lung Cancer: Yes | Lung Cancer: No | Total |
|---------------------|-------------------|-----------------|-------|
| Smoker              | 60                | 140             | 200   |
| Non-smoker          | 30                | 170             | 200   |
| **Total**           | 90                | 310             | 400   |

##### 3. Calculate the Expected Frequencies
The expected frequency for each cell is calculated using the formula:

$ E_{ij} = \frac{( \text{Row Total}_i \times \text{Column Total}_j )}{\text{Grand Total}} $

For each cell:

- **Expected frequency for Smoker & Lung Cancer: Yes:** 
  $ E_{11} = \frac{200 \times 90}{400} = 45 $
- **Expected frequency for Smoker & Lung Cancer: No:** 
  $ E_{12} = \frac{200 \times 310}{400} = 155 $
- **Expected frequency for Non-smoker & Lung Cancer: Yes:** 
  $ E_{21} = \frac{200 \times 90}{400} = 45 $
- **Expected frequency for Non-smoker & Lung Cancer: No:** 
  $ E_{22} = \frac{200 \times 310}{400} = 155 $

##### 4. Compute the Chi-Square Statistic
The chi-square statistic is calculated using:

$ \chi^2 = \sum \frac{( O_{ij} - E_{ij} )^2}{E_{ij}} $

where $ O_{ij} $ is the observed frequency, and $ E_{ij} $ is the expected frequency.

Calculate for each cell:

- **For Smoker & Lung Cancer: Yes:** 
  $ \frac{(60 - 45)^2}{45} = \frac{225}{45} = 5 $
- **For Smoker & Lung Cancer: No:** 
  $ \frac{(140 - 155)^2}{155} = \frac{225}{155} \approx 1.452 $
- **For Non-smoker & Lung Cancer: Yes:** 
  $ \frac{(30 - 45)^2}{45} = \frac{225}{45} = 5 $
- **For Non-smoker & Lung Cancer: No:** 
  $ \frac{(170 - 155)^2}{155} = \frac{225}{155} \approx 1.452 $

Sum of all these values:

$ \chi^2 = 5 + 1.452 + 5 + 1.452 = 12.904 $

##### 5. Determine the Degrees of Freedom
Degrees of freedom for a chi-square test in a contingency table is:

$ \text{df} = (r - 1) \times (c - 1) $

where $ r $ is the number of rows, and $ c $ is the number of columns.

For this table:

$ \text{df} = (2 - 1) \times (2 - 1) = 1 $

##### 6. Compare with the Critical Value
At a significance level of 0.05 and 1 degree of freedom, the critical value from the chi-square distribution table is approximately 3.841.

##### 7. Conclusion
- **Calculated Chi-Square Statistic:** 12.904
- **Critical Value:** 3.841

Since 12.904 > 3.841, we reject the null hypothesis.

##### Interpretation
There is a significant association between smoking status and lung cancer diagnosis at the 0.05 significance level.

- **Alternative Hypothesis (H‚ÇÅ)**: There is an association between chocolate preference and country of origin (i.e., chocolate preference is not independent of the country).

##### 2. Calculate the Expected Frequencies

First, compute the expected frequencies for each cell in the contingency table under the assumption that the null hypothesis is true.

To find the expected frequency for each cell, use the formula:

$$
E_{ij} = \frac{( \text{Row Total}_i ) \times ( \text{Column Total}_j )}{\text{Grand Total}}
$$

where:

- $ E_{ij} $ = Expected frequency for cell (i, j)
- $ \text{Row Total}_i $ = Total count for row i
- $ \text{Column Total}_j $ = Total count for column j
- $ \text{Grand Total} $ = Total count of all observations

**Totals:**

- Total for Milk Chocolate = 200 (U.S.) + 225 (U.K.) = 425
- Total for Dark Chocolate = 150 (U.S.) + 175 (U.K.) = 325
- Total for White Chocolate = 150 (U.S.) + 100 (U.K.) = 250
- Total for U.S. = 500
- Total for U.K. = 500
- Grand Total = 1000

**Expected Frequencies:**

- For Milk Chocolate (U.S.): 
  $$
  E_{11} = \frac{500 \times 425}{1000} = 212.5
  $$

- For Dark Chocolate (U.S.): 
  $$
  E_{12} = \frac{500 \times 325}{1000} = 162.5
  $$

- For White Chocolate (U.S.): 
  $$
  E_{13} = \frac{500 \times 250}{1000} = 125
  $$

- For Milk Chocolate (U.K.): 
  $$
  E_{21} = \frac{500 \times 425}{1000} = 212.5
  $$

- For Dark Chocolate (U.K.): 
  $$
  E_{22} = \frac{500 \times 325}{1000} = 162.5
  $$

- For White Chocolate (U.K.): 
  $$
  E_{23} = \frac{500 \times 250}{1000} = 125
  $$

##### 3. Compute the Chi-Square Statistic

The chi-square statistic is calculated using the formula:

$$
\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
$$

where:

- $ O_{ij} $ = Observed frequency for cell (i, j)
- $ E_{ij} $ = Expected frequency for cell (i, j)

**Calculations:**

- For Milk Chocolate (U.S.): 
  $$
  \frac{(200 - 212.5)^2}{212.5} = \frac{156.25}{212.5} \approx 0.735
  $$

- For Dark Chocolate (U.S.): 
  $$
  \frac{(150 - 162.5)^2}{162.5} = \frac{156.25}{162.5} \approx 0.961
  $$

- For White Chocolate (U.S.): 
  $$
  \frac{(150 - 125)^2}{125} = \frac{625}{125} = 5.0
  $$

- For Milk Chocolate (U.K.): 
  $$
  \frac{(225 - 212.5)^2}{212.5} = \frac{156.25}{212.5} \approx 0.735
  $$

- For Dark Chocolate (U.K.): 
  $$
  \frac{(175 - 162.5)^2}{162.5} = \frac{156.25}{162.5} \approx 0.961
  $$

- For White Chocolate (U.K.): 
  $$
  \frac{(100 - 125)^2}{125} = \frac{625}{125} = 5.0
  $$

Summing these values gives the chi-square statistic:

$$
\chi^2 = 0.735 + 0.961 + 5.0 + 0.735 + 0.961 + 5.0 \approx 13.392
$$

##### 4. Determine the Degrees of Freedom

Degrees of freedom (df) for a chi-square test of independence is given by:

$$
df = (R - 1) \times (C - 1)
$$

where $ R $ is the number of rows and $ C $ is the number of columns.

In this case, $ R = 2 $ (U.S. and U.K.) and $ C = 3 $ (Milk Chocolate, Dark Chocolate, White Chocolate):

$$
df = (2 - 1) \times (3 - 1) = 1 \times 2 = 2
$$

##### 5. Find the Critical Value and Compare

Using a chi-square distribution table, find the critical value for $ df = 2 $ at the 0.01 significance level. The critical value is approximately 9.210.

##### 6. Make a Decision

Compare the chi-square statistic to the critical value:

- If $ \chi^2 > \text{critical value} $, reject the null hypothesis.
- If $ \chi^2 \leq \text{critical value} $, fail to reject the null hypothesis.

Here, $ \chi^2 = 13.392 $ and the critical value is 9.210. Since $ 13.392 > 9.210 $, we reject the null hypothesis.

##### Conclusion

There is sufficient evidence at the 0.01 significance level to conclude that there is a significant association between chocolate preference and country of origin.


#### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

### One-Sample t-Test to Test if the Population Mean is Significantly Different from 70

#### 1. State the Hypotheses
- **Null Hypothesis (H‚ÇÄ):** The population mean is 70. 
  $$ H_0: \mu = 70 $$
- **Alternative Hypothesis (H‚ÇÅ):** The population mean is not 70.
  $$ H_1: \mu \neq 70 $$

#### 2. Gather the Sample Data
- Sample mean ($\bar{x}$) = 72
- Sample standard deviation (s) = 10
- Sample size (n) = 30
- Significance level ($\alpha$) = 0.05

#### 3. Calculate the Test Statistic
Since the population standard deviation is unknown, we use the t-test statistic:

$$ t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}} $$

Where:
- $\bar{x}$ = sample mean = 72
- $\mu_0$ = population mean under null hypothesis = 70
- $s$ = sample standard deviation = 10
- $n$ = sample size = 30

So:

$$ t = \frac{72 - 70}{10 / \sqrt{30}} \approx \frac{2}{10 / 5.477} \approx \frac{2}{1.825} \approx 1.095 $$

#### 4. Determine the Critical t-Value
For a two-tailed test with $\alpha = 0.05$ and $n - 1 = 29$ degrees of freedom, you can find the critical t-value from the t-distribution table or using statistical software.

The critical t-value for $\alpha = 0.05$ (two-tailed) and $df = 29$ is approximately $\pm 2.045$.

#### 5. Make the Decision
Compare the calculated t-value to the critical t-value:

- Calculated t-value: 1.095
- Critical t-values: $\pm 2.045$

Since 1.095 is less than 2.045, we fail to reject the null hypothesis.

#### 6. Conclusion
At the 0.05 significance level, there is not enough evidence to conclude that the population mean is significantly different from 70.
