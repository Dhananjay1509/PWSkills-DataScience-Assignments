Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web framework for Python, designed to facilitate quick and easy web application development. It is classified as a microframework, meaning it has minimal dependencies and provides the essential tools needed to build web applications without unnecessary complexity. Flask was developed by Armin Ronacher and is built on the WSGI (Web Server Gateway Interface) toolkit and the Jinja2 template engine, allowing developers to create dynamic web pages efficiently.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to application endpoints, which are functions that handle HTTP requests and return responses. In other words, routing is about defining how the URL of a request should be handled by the application.

We use app routes for several reasons:

1) Url matching: Routes allow us to match incoming requests to specific URLs, enabling the application to respond correctly to different requests.
2) Function binding: Routes associate a URL with a specific function, which is responsible for handling the request. This allows us to write reusable code and avoid repetition.
3) Control flow: Routes provide a way to control the flow of the application, directing requests to the appropriate functions based on the URL.
4) HTTP methods support: Routes can specify the HTTP method (e.g., GET, POST, PUT, DELETE) that a function can handle, allowing for more robust handling of different types of requests.
5) Decorators: Routes can use decorators, which are lightweight functions that can modify the behavior of the function, such as logging or authentication.
6) URL parameters: Routes can capture URL parameters, such as IDs or query strings, and pass them as arguments to the associated function.
7) Error handling: Routes provide a way to handle errors and exceptions, ensuring that the application responds correctly to unexpected situations.
   

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook.


```python
from flask import Flask, render_template_string

app = Flask(__name__)


@app.route("/")
@app.route("/welcome")
def index():
    template = """
    <html>
        <body>
            <h1>Welcome to ABC Corporation</h1>
            <h2>Company Name: ABC Corporation</h2>
            <h2>Location: India</h2>
            <h2>Contact Detail: 999-999-9999</h2>
        </body>
    </html>
    """
    return render_template_string(template)


if __name__ == "__main__":
    app.run(host="127.0.0.2", port=5000, debug=True)
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function in Flask is used to generate a URL to a function based on the view function's name and any variable rules.

Here's an example of how it works:

```python
from flask import Flask, url_for

app = Flask(__name__)


@app.route("/")
def home():
    return "Welcome to Home Page"


@app.route("/user/<username>")
def show_user_profile(username):
    return "User %s" % username


@app.route("/post/<int:post_id>")
def show_post(post_id):
    return "Post %d" % post_id


@app.route("/path/<path:subpath>")
def show_subpath(subpath):
    return "Subpath %s" % subpath


@app.route("/generate-urls")
def generate_urls():
    cursor = [(1, "Hello"), (2, "World"), (3, "Python")]
    urls = []

    for id, name in cursor:
        url = url_for("show_post", post_id=id)
        urls.append(url)

    user_url = url_for("show_user_profile", username="John")
    subpath_url = url_for("show_subpath", subpath="/users/me")

    return {
        "post_urls": urls,
        "user_url": user_url,
        "subpath_url": subpath_url,
    }


if __name__ == "__main__":
    app.run(host="127.0.0.1", port=5000, debug=True)
```
